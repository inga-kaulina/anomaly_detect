```python
#ik18075 for anomaly detection
# 1 day csv data, to print out anomalies detected
#import necessary modules
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn import preprocessing
import pandas as pd
import gensim as gs
import sys
import numpy
numpy.set_printoptions(threshold=sys.maxsize)
```

```python
data_csv = pd.read_csv('C:/salaspils.csv', error_bad_lines=False) #import data
data_original = data_csv

# create the Labelencoder object
le = preprocessing.LabelEncoder()

data_csv['ALOG_INFO'] = le.fit_transform(data_csv['ALOG_INFO']) #transform data to vector
data_csv['FRQ'] = data_csv['ALOG_INFO'].map(data_csv['ALOG_INFO'].value_counts())

cols = [col for col in data_csv.columns if col in ['ID','ALOG_INFO']] 
data = data_csv[cols]
target = data_csv['FRQ'] #for each class add weight
```

```python
from sklearn.model_selection import train_test_split #split data 4 ways
data_train, data_test, target_train, target_test = train_test_split(data, target, test_size = 0.30, random_state = 10)
```

```python
# import the necessary module
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

neigh = KNeighborsClassifier(n_neighbors=1) #create machine learning process based on KNN (k-nearest neighbours)

neigh.fit(data_train, target_train) #model trained based on data_train and target_train

pred = neigh.predict(data_test) # split data in data_test is checked against trained model, to find anomalies

#convert to dataframe using Pandas
an_score = pd.DataFrame(data=data_test)

#add anomaly calculation
an_score['Anomality_score'] = pred
result = an_score.loc[an_score['Anomality_score'] <= 1]

ids = result['ID']
ids_arr = ids.to_numpy()

#filter out anomalies from original full set of data
return_result = pd.DataFrame()
for id_arr in ids_arr:
    added = data_original.loc[data_original['ID'] == id_arr]
    return_result = return_result.append(added)
    
#result csv with anomalies
return_result.to_csv('C:/scikit/anomalies.csv', index = False)
print(return_result)

#print the accuracy score of the model
#print ("KNeighbors accuracy score : ",accuracy_score(target_test, pred))
```
